```
import pandas as pd
import folium
```

Read datasets with pandas

```
agriculture_kenya = pd.read_csv('land_under_Irrigation.csv')
california_housing = pd.read_csv('sample_data/california_housing_test.csv')
energy_kenya = pd.read_csv('Main_Energy_Sources.csv')
```

```
# Initialze The kenyan map
kenya = folium.Map([-0.023559, 37.9061928], zoom_start=8)

kenya
```

```
# initialize map of carlifonia
california = folium.Map([37.229564, -120.047533], zoom_start=8)
california
```

```
#Decide what we wish to visualize from the given CSVs onto the map and make a Geo-Centric visualization
# Visualizations will be my means of bubbles whose size is proportional to the quantity being represented
#For Kenya

# preprocess locations to float values

agriculture_kenya.drop(agriculture_kenya.tail(1).index, inplace=True)

for index, row in agriculture_kenya.iterrows():
  #print(float(str(row['Location_1']).split(",")[0][1:]), float(str(row['Location_1']).split(",")[1][:-1]) )
  if float(str(row['Percent']).split('%')[0]) >= 50:
    marker_color = 'green'
    fill = 'green'
  elif float(str(row['Percent']).split('%')[0]) >= 30:
    marker_color = 'orange'
    fill = 'orange'
  elif float(str(row['Percent']).split('%')[0]) >= 20:
    marker_color = 'yellow'
    fill = 'yellow'
  elif float(str(row['Percent']).split('%')[0]) >= 10:
    marker_color = 'darkred'
    fill = 'darkred'
  else:
    marker_color='darkred'
    fill = 'darkpurple'
        

    folium.Circle(
          location=[float(str(row['Location_1']).split(",")[0][1:]), float(str(row['Location_1']).split(",")[1][:-1]) ],
          popup= str(row['Percent'])+' of agricultural land under irrigation',
          tooltip=row['County'],
          radius=float(str(row['Total_Count']).split('%')[0])/10,
          color=marker_color,
          fill=True,
          fill_color=fill,
          
       ).add_to(kenya)
    
    
  
kenya
```
